### Want the average number of episodes per season

## How many years did the show run for?

```python
def years_run(row):
    count = row['End_Year'] - row['Start_Year']
    
    if count < 0:
        count = 2020 - row['Start_Year']
    
    if count == 0:
        count = 1

    return count

tv_show_df['Years_Ran'] = tv_show_df.apply(years_run, axis=1)
```

### Now fix the avg num of episodes per season for our outliers

```python
#define a function that gets the avg num of episodes per season for a specific row

def avg_num_eps(row_index):
    #divide number of episodes by number of seasons
    tv_show_df.loc[row_index,'Num_Episodes_Per_Season'] = tv_show_df.loc[row_index]['Num_of_Episodes'] / tv_show_df.loc[row_index]['Num_of_Seasons']

    #round it off to 2 decimal places
    tv_show_df.loc[row_index,'Num_Episodes_Per_Season'] = tv_show_df.loc[row_index,'Num_Episodes_Per_Season'].round(2)

    tv_show_df.loc[row_index]
```

## How many unique genres do we have?
### Below is in the original dataframe before I fixed the outliers

```python
#find how many unique genres
from collections import defaultdict

genre_count = defaultdict(int)

for genre in tv_show_df['Genres']:
    for i in range(len(genre)):
        genre_count[genre[i]] += 1
    
genre_count
```

```python
len(genre_count)
```

### We have some genres that show up way less frequently, maybe we can put them in an 'Other' genre

```python
filtered_gens = {'Other':0}
for genre in genre_count:
    if genre_count[genre] > 7:
        filtered_gens[genre] = genre_count[genre]
    else:
        filtered_gens['Other'] += genre_count[genre]
        print(genre, genre_count[genre])
```

```python
filtered_gens
```

```python
len(filtered_gens)
```

```python
#replace the "others" in the df
others_genre_list = []
for genre_list in tv_show_df['Genres']:
    temp = []
    for genre in genre_list:
        if genre in filtered_gens:
            temp.append(genre)

        else:
            temp.append('Other')
    others_genre_list.append(temp)
    
len(others_genre_list)
```

```python
def drop_row_by_index(df, title):
    #if there are multiple idx's that this happens at, then idx returns a list
    idx = df[(df.Title == title)].index
    
    for j in idx:
        print(j)
        df.drop(j, inplace = True)
```

### Now want to add a column with ordinal values for the MPAA rating column and drop the None values

```python
#check all the possible MPAA ratings a show in our df can have
inner_merge_dummy_genre_MPAA['MPAA_Rating'].unique()
```

```python
#create a dictionary to use as an ordinal mapping
ordinal_map = {'TV-Y':0, 'TV-Y7': 1, 'TV-Y7-FV':2, 
               'TV-G':3, 'TV-PG':4, 'TV-14':5, 'TV-MA':6}
```

```python
#make a duplicate row
inner_merge_dummy_genre_MPAA['Ordinal_MPAA'] = inner_merge_dummy_genre_MPAA['MPAA_Rating']
```

```python
#fill in the values using the ordinal mapping dict
for idx, rate in enumerate(inner_merge_dummy_genre_MPAA['Ordinal_MPAA']):
    try:
        inner_merge_dummy_genre_MPAA.loc[idx,'Ordinal_MPAA'] = ordinal_map[rate]
        
    #some rows say none, so we'll drop those rows
    except:
        inner_merge_dummy_genre_MPAA.drop(idx, inplace = True)
    
```
